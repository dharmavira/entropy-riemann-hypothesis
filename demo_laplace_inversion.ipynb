# Laplace Inversion Demo

Recover zeta zeros from entropy trace.

```python
import numpy as np
import matplotlib.pyplot as plt
from entropy_trace.entropy_model import theta
from entropy_trace.laplace_inversion import recover_zeros

# Known gamma values (first 10)
gamma_true = np.array([
    14.134725, 21.022040, 25.010858, 30.424876, 32.935061,
    37.586178, 40.918719, 43.327073, 48.005150, 49.773832
])

t_values = np.linspace(0.001, 0.1, 100)
theta_values = np.array([theta(t, gamma_true) for t in t_values])

# Recover zeros
gamma_est = recover_zeros(t_values, theta_values, N=10)

print("Recovered γ_n:")
print(gamma_est)
```
